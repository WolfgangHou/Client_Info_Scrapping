# 使用Selenium爬取網站公開email

```python
# 在本地端運行Jupyter Notebook
# 依照list的url進行爬取email：for迴圈
# 準備爬取
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import re

# 創立新的資料框儲存爬取資訊
df_begin = pd.DataFrame()
df_begin['brand'] = []
df_begin['title'] = []
df_begin['descrip'] = []
df_begin['url'] = []
df_begin['email'] = []





for i in df_scrape_less['URL']:
  try:
    # 訪問待搜索網頁
    # 創建WebDriver
    options = Options()
    options.add_argument('--disable-blink-features=AutomationControlled')
    options.add_argument("--disable-extensions")
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    # Set custom headers
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    options.add_argument(f"--window-size=1920,1080")
    options.add_argument("--start-maximized")
    driver = webdriver.Chrome(options=options)
    driver.execute_script(
        "Object.defineProperty(navigator, 'userAgent', {value: arguments[0], writable: false, configurable: true});",
        headers["User-Agent"])
    driver.get(i)

    # 取得網頁原始碼
    content = driver.page_source

    # 定義正則表達式用於匹配電子郵件地址
    email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    email_dict = {}
    # 在網頁原始碼中搜尋所有可能的郵件地址並儲存在email_dict之中
    for match in re.findall(email_pattern, content): #在網頁內容中找尋email
        if driver.current_url not in email_dict: #確認現在瀏覽器的網址並未存在於email_dict中
            email_dict[driver.current_url] = [] #建立以目前網址為key的字典項目
        if match not in email_dict[driver.current_url]: #確認在目前url搜尋到的email是否已經在key中有對應的值 如果沒有
            email_dict[driver.current_url].append(match) #將match的value附加到key對應的value之中
        #將字典轉換成資料框
        data = []
        for url, emails in email_dict.items():
            for email in emails:
                # 將每個email都補上其他欄位的對應資訊後新增到名為data
                title = df_scrape_less[df_scrape_less['URL'] == url]['Title'].values[0]
                descrip = df_scrape_less[df_scrape_less['URL'] == url]['Description'].values[0]
                brand = df_scrape_less[df_scrape_less['URL'] == url]['brandName'].values[0]
                data.append([url, email, title, descrip, brand])

        df_new = pd.DataFrame(data, columns=['url', 'email', 'title', 'descrip', 'brand'])
    df_end = pd.concat([df_begin, df_new], axis = 1)
    print(df_end)
    df_end.to_csv('email爬取結果.csv')



  except InvalidSessionIdException as e:
    print("InvalidSessionIdException")
  finally:
    # 關閉瀏覽器窗口
    driver.quit()

```
